```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_df = mouse_drug_df.merge(clinical_trial_df, on="Mouse ID")

# Display the data table for preview
combined_df.head()


```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_volume = combined_df[['Drug', 'Timepoint', 'Tumor Volume (mm3)']].groupby(['Drug', 'Timepoint']).mean()
# Convert to DataFrame
mean_tumor_volume.reset_index(inplace=True)
# Preview DataFrame
mean_tumor_volume.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard_error = combined_df[['Drug', 'Timepoint', 'Tumor Volume (mm3)']].groupby(['Drug', 'Timepoint']).sem()
# Convert to DataFrame
standard_error.reset_index(inplace=True)
# Preview DataFrame
standard_error.head()
```

```python
std_error_pivot = pd.pivot_table(standard_error, values='Tumor Volume (mm3)', index='Timepoint', columns='Drug')
std_error_pivot.reset_index(inplace=True)
std_error_pivot.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tumor_pivot = pd.pivot_table(mean_tumor_volume, values='Tumor Volume (mm3)', index='Timepoint', columns='Drug')
# Preview that Reformatting worked
drug_names = mean_tumor_pivot.columns.tolist()
mean_tumor_pivot.reset_index(inplace=True)
timepoints = mean_tumor_pivot['Timepoint'].tolist()
mean_tumor_pivot.head()
```

```python
#colors = ['red', 'green', 'blue', 'orange', 'purple', 'black', 'yellow', 'gray', 'DarkRed', 'LightGreen']
```

```python

# Generate the Plot (with Error Bars)
for drugs in drug_names:
    plt.scatter(x=timepoints, y=mean_tumor_pivot[drugs])
    plt.legend(bbox_to_anchor=(1.05, 1.02), loc=2, fontsize=11)
    
for drugs in drug_names:
    plt.errorbar(x=timepoints, y=mean_tumor_pivot[drugs], yerr=std_error_pivot[drugs])
#plt.legend(loc=2, fontsize=7.8)
plt.xlabel('Time')
plt.ylabel('Tumor Volume (mm3)') 
plt.title('Tumor Volume on Different Drugs by Time')
plt.grid()
# Save the Figure

```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_site = combined_df[['Drug', 'Timepoint', 'Metastatic Sites']].groupby(['Drug', 'Timepoint']).mean()
# Convert to DataFrame
mean_met_site.reset_index(inplace=True)
# Preview DataFrame
mean_met_site.head()

```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
standard_error_met = combined_df[['Drug', 'Timepoint', 'Metastatic Sites']].groupby(['Drug', 'Timepoint']).sem()
# Convert to DataFrame
standard_error_met.reset_index(inplace=True)
# Preview DataFrame
standard_error_met.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_met_pivot = pd.pivot_table(mean_met_site, values='Metastatic Sites', index='Timepoint', columns='Drug')
# Preview that Reformatting worked
drug_names = mean_met_pivot.columns.tolist()
mean_met_pivot.reset_index(inplace=True)
timepoints = mean_met_pivot['Timepoint'].tolist()
mean_met_pivot.head()
```

```python
std_error_pivot_1 = pd.pivot_table(standard_error_met, values='Metastatic Sites', index='Timepoint', columns='Drug')
std_error_pivot_1.reset_index(inplace=True)
std_error_pivot_1.head()
```

```python
# Generate the Plot (with Error Bars)
for drugs in drug_names:
    plt.scatter(x=timepoints, y=mean_met_pivot[drugs])
    plt.legend(bbox_to_anchor=(1.05, 1.02), loc=2, fontsize=11)
    
for drugs in drug_names:
    plt.errorbar(x=timepoints, y=mean_met_pivot[drugs], yerr=std_error_pivot_1[drugs])
#plt.legend(loc=2, fontsize=7.8)
plt.xlabel('Time')
plt.ylabel('Metastatic Sites') 
plt.title('Average Metastatic Sites on Different Drugs by Time')
plt.grid()
# Save the Figure

# Show the Figure
```
