Insights: 
1. Most drugs are not much more effective than the placebo
2. Most drugs didn't decrease the tumor volume and killed more than half the mice
3. There was an overall increase in number of metastatic sites regardless of drug and some mice always died, so none of the drugs was fool proof


```python


# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_df = mouse_drug_df.merge(clinical_trial_df, on="Mouse ID")

# Display the data table for preview
combined_df.head()


```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_volume = combined_df[['Drug', 'Timepoint', 'Tumor Volume (mm3)']].groupby(['Drug', 'Timepoint']).mean()
# Convert to DataFrame
mean_tumor_volume.reset_index(inplace=True)
# Preview DataFrame
mean_tumor_volume.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard_error = combined_df[['Drug', 'Timepoint', 'Tumor Volume (mm3)']].groupby(['Drug', 'Timepoint']).sem()
# Convert to DataFrame
standard_error.reset_index(inplace=True)
# Preview DataFrame
standard_error.head()
```

```python
std_error_pivot = pd.pivot_table(standard_error, values='Tumor Volume (mm3)', index='Timepoint', columns='Drug')
std_error_pivot.reset_index(inplace=True)
std_error_pivot.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tumor_pivot = pd.pivot_table(mean_tumor_volume, values='Tumor Volume (mm3)', index='Timepoint', columns='Drug')
# Preview that Reformatting worked
drug_names = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']
colors = ['red', 'blue', 'green', 'black']
markers = ['D', '^', 's', 'd']
mean_tumor_pivot.reset_index(inplace=True)
timepoints = mean_tumor_pivot['Timepoint'].tolist()
mean_tumor_pivot.head()
```

```python
#colors = ['red', 'green', 'blue', 'orange', 'purple', 'black', 'yellow', 'gray', 'DarkRed', 'LightGreen']
```

```python

# Generate the Plot (with Error Bars)
for drugs, color, marker in zip(drug_names, colors, markers):
        plt.scatter(x=timepoints, y=mean_tumor_pivot[drugs], color=color, marker=marker)
        plt.legend()
    
for drugs, color in zip(drug_names, colors):
    plt.errorbar(x=timepoints, y=mean_tumor_pivot[drugs], yerr=std_error_pivot[drugs], color=color)
#plt.legend(loc=2, fontsize=7.8)
plt.xlabel('Time')
plt.ylabel('Tumor Volume (mm3)') 
plt.title('Tumor Volume on Different Drugs by Time')
plt.grid()
# Save the Figure
plt.savefig('Tumor Volume on Different Drugs by Time.png')

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_site = combined_df[['Drug', 'Timepoint', 'Metastatic Sites']].groupby(['Drug', 'Timepoint']).mean()
# Convert to DataFrame
mean_met_site.reset_index(inplace=True)
# Preview DataFrame
mean_met_site.head()

```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
standard_error_met = combined_df[['Drug', 'Timepoint', 'Metastatic Sites']].groupby(['Drug', 'Timepoint']).sem()
# Convert to DataFrame
standard_error_met.reset_index(inplace=True)
# Preview DataFrame
standard_error_met.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_met_pivot = pd.pivot_table(mean_met_site, values='Metastatic Sites', index='Timepoint', columns='Drug')
# Preview that Reformatting worked
#drug_names = mean_met_pivot.columns.tolist()
mean_met_pivot.reset_index(inplace=True)
timepoints = mean_met_pivot['Timepoint'].tolist()
mean_met_pivot.head()
```

```python
std_error_pivot_1 = pd.pivot_table(standard_error_met, values='Metastatic Sites', index='Timepoint', columns='Drug')
std_error_pivot_1.reset_index(inplace=True)
std_error_pivot_1.head()
```

```python
# Generate the Plot (with Error Bars)
for drugs, color, marker in zip(drug_names, colors, markers):
    plt.scatter(x=timepoints, y=mean_met_pivot[drugs], color=color, marker=marker)
    plt.legend()
    
for drugs, color in zip(drug_names, colors):
    plt.errorbar(x=timepoints, y=mean_met_pivot[drugs], yerr=std_error_pivot_1[drugs], color=color)
#plt.legend(loc=2, fontsize=7.8)
plt.xlabel('Time')
plt.ylabel('Metastatic Sites') 
plt.title('Average Metastatic Sites on Different Drugs by Time')
plt.grid()
# Save the Figure
plt.savefig('Average Metastatic Sites on Different Drugs by Time')
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count = combined_df[['Drug', 'Timepoint', 'Mouse ID']].groupby(['Drug', 'Timepoint']).count()
# Convert to DataFrame
mouse_count.reset_index(inplace=True)
mouse_count = mouse_count.rename(columns={'Mouse ID': "Mouse Count"})
# Preview DataFrame
mouse_count.head()
```

```python


# Minor Data Munging to Re-Format the Data Frames
mouse_count_pivot = pd.pivot_table(mouse_count, values='Mouse Count', index='Timepoint', columns='Drug')
# Preview that Reformatting worked
#drug_names = mouse_count_pivot.columns.tolist()
mouse_count_pivot.reset_index(inplace=True)
timepoints = mouse_count_pivot['Timepoint'].tolist()
mouse_count_pivot = mouse_count_pivot.set_index("Timepoint")
```

```python
mouse_pct_survival = mouse_count_pivot.div(mouse_count_pivot.loc[0], axis='columns')*100
```

```python
# Generate the Plot (Accounting for percentages)
for drugs, color, marker in zip(drug_names, colors, markers):
    plt.scatter(x=timepoints, y=mouse_pct_survival[drugs], color=color, marker=marker)
    plt.legend()
    
for drugs, color in zip(drug_names, colors):
    plt.plot(timepoints, mouse_pct_survival[drugs], color=color)
plt.xlabel('Time')
plt.ylabel('Percent of Mice Alive') 
plt.title('Survival Rate of Mice by Drug and Time')
plt.grid()
# Save the Figure
plt.savefig('Survival Rate of Mice by Drug and Time')
# Show the Figure
plt.show()
```
